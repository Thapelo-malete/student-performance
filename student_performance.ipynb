# Student Performance

```python
import pandas as pd

data = pd.read_csv("data/student_performance.csv")
data.head()
```

```python
data.columns
```

```python
new_columns = []

for c in data.columns:
    new_columns.append("_".join(c.lower().split()))

new_columns
```

```python
data.columns = new_columns
data
```

```python
X, y = data.iloc[:, :-1], data.iloc[:,-1]
X.head(2)
```

```python
y.head(2)
```

```python
from sklearn.preprocessing import OneHotEncoder


encoder = OneHotEncoder(drop='first')
encoder.fit(pd.DataFrame(X.extracurricular_activities))
X = X.join(pd.DataFrame(encoder.transform(pd.DataFrame(X.extracurricular_activities)).toarray(), columns=encoder.get_feature_names_out(), index=X.index)).drop(columns=["extracurricular_activities"])
X
```

```python
X.rename(columns={'extracurricular_activities_Yes': 'extracurricular_activities' }, inplace=True)
X
```

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=5)
X_train
```

```python
from sklearn.linear_model import LinearRegression


model = LinearRegression()
model.fit(X_train, y_train)

predictions = model.predict(X_test)
predictions
```

```python
from sklearn.metrics import r2_score


print(f"RÂ² : {r2_score(y_test, predictions)}")
```
